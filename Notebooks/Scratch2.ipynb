```python
### Now compare to the result of my old method, which works (but is too memory intensive for the full dataset). 
# This represents the final output I'm looking for 

indexes = ('item_id','date')
full = pd.read_table('../Results/ent.tsv',header=None,names=colNames,na_values=["\N"],index_col=indexes,parse_dates='date',nrows=10000)
full = full.drop('seq',1) # 'seq' column as currently defined doesn't apply here; drop it
 # this function reindexes each item so we have rows for every month from the first observation to 'maxDate' (for that item)
def switchToSeqIndex(df):
    minDate = df.index[0][1] # get the first observed date
    return df.reset_index(level='item_id',drop=True).reset_index().set_index('date') \
            .reindex(pd.date_range(minDate,maxDate,freq='MS'),method='ffill').reset_index('date',drop=True)
full = full.groupby(level='item_id').apply(switchToSeqIndex)
full.index.names[1]='seq' # now we can rename our new 'seq' index

full[350:400]
```
