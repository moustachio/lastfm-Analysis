```python
import pandas as pd
import numpy as np
import os

maxDate = '2012-12-01' # Don't look at any date past December 2012 
rows = 100000 # For debugging, only read in this many rows (set to None for full date)
colNames = ['item_id','date','ent_cumulative','rel_ent_cumulative','gini_cumulative','tags_cumulative','anno_cumulative','ent_current','rel_ent_current','gini_current','tags_current','anno_current','topCopy','binCopy','normCopy','seq']

# Build the "full" dataframe, which *does* include months where there were no observations 
indexes = ('item_id','date')
full = pd.read_table('../Results/ent.tsv',header=None,names=colNames,na_values=["\N"],index_col=indexes,parse_dates='date',nrows=rows)
full = full.drop('seq',1) # 'seq' column as currently defined doesn't apply here; drop it
```

```python
#store.select_column('full','item_id').unique()
#test = store.select('full', [pd.Term('item_id', '==', 190)])
#x = store['expanded'].reset_index().set_index(['item_id','index'])
#x[150:200]
#x.mean(level='item_id').head()
#x.index.name
#store
#full[0:1]
import pandas as pd
import dateutil
maxDate = '2012-12-01' # Don't look at any date past December 2012 
colNames = ['item_id','date','ent_cumulative','rel_ent_cumulative','gini_cumulative','tags_cumulative','anno_cumulative','ent_current','rel_ent_current','gini_current','tags_current','anno_current','topCopy','binCopy','normCopy','seq']
full = pd.read_table('../Results/ent.tsv',header=None,names=colNames,na_values=["\N"],converters={'date':dateutil.parser.parse},nrows=10000)
full = full.drop('seq',1) # 'seq' column as currently defined doesn't apply here; drop it
dates_df = pd.DataFrame()
for i in full.groupby('item_id').apply(lambda x: x[0:1]).date:
    rng = pd.date_range(i,maxDate,freq='MS')
    idx = range(len(rng))
    dates_df = dates_df.append(pd.DataFrame({'date':rng,'seq':idx}))
```
