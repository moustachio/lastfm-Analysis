```python
### Tag stabilization - item

item=690 # 690=Radiohead

fig, ax = plt.subplots(nrows=1, ncols=1) 
ax.set_color_cycle(['r', 'g', 'b', 'y','k','m','c'])
itemData = data.xs(item,level='iid').sum(level=['tag','date'])
totalAnnoByMonth = itemData.sum(level='date').cumsum().reindex(baseDateRange,method='ffill')
sortedTagFreqs = itemData.sum(level='tag').sort('anno',ascending=False).index
legendLabels = []
nLabels=5
for i,tag in enumerate(sortedTagFreqs[:20]):
    if i<nLabels:
        legendLabels.append(tag)
    (itemData.xs(tag,level='tag').cumsum().reindex(baseDateRange,method='ffill') / totalAnnoByMonth).plot(ax=ax,legend=False)

ax.set_ylabel("Relative Tag Proportion")
handles,labels = ax.get_legend_handles_labels()
ax.legend(handles[:5],legendLabels,loc='upper left')
```

```python
### Tag stabilization -- GLOBAL

fig, ax = plt.subplots(nrows=1, ncols=1)
ax.set_color_cycle(['r', 'g', 'b', 'y','k','m','c'])
plotData = data.sum(level=['tag','date'])
totalAnnoByMonth = plotData.sum(level='date').cumsum().reindex(baseDateRange,method='ffill')
sortedTagFreqs = plotData.sum(level='tag').sort('anno',ascending=False).index
legendLabels = []
nLabels=5
for i,tag in enumerate(sortedTagFreqs[:20]):
    if i<nLabels:
        legendLabels.append(tag)
    (plotData.xs(tag,level='tag').cumsum().reindex(baseDateRange,method='ffill') / totalAnnoByMonth).plot(ax=ax,legend=False)
    
ax.set_ylabel("Relative Tag Proportion")
handles,labels = ax.get_legend_handles_labels()
ax.legend(handles[:nLabels],legendLabels,loc='upper right')
```
