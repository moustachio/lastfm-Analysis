Import modules and load data
---

```python
import pandas as pd
pylab.rcParams['figure.figsize'] = (12.0, 8.0)

d = '../results/generalData/'

globalTagRanks = pd.read_table(d+'annotationsPerTag',header=None,names=('tag','freq'),index_col='tag')
globalTagRanks = globalTagRanks.rank(ascending=False).reset_index()
globalTagRanks.columns = ('tag','rank')

annoPerUser = pd.read_table(d+'annotationsPerUser',header=None,names=('uid','freq'),index_col='uid')

allAnnotations = pd.read_table('../Results/annotations_date-iid-tid.tsv',header=None,names=('uid','iid','artist','tag','date')) # Read in .tsv
```

---
Average Tag Rank vs. Total Annotations
---

Each point corresponds to the average tag rank across users with x annotations.

* x-axis: total annotations
* y-axis: Mean global tag rank (i.e. tag rank of 5 refers to the 5th most popular tag *overall*) 


```python
merged = allAnnotations.merge(globalTagRanks,on='tag',how='left') # get corresponding tag rank for each annoations
avTagRank = merged.groupby('uid').rank.mean() # groupby user_id to get the mean tag rank for each user
annoVsTagRanks = annoPerUser.join(avTagRank)# put each user's number of annotations and mean tag rank in the same dataframe
```

```python
# Now calculate averages and plot
annoVsTagRanks.columns = ('annotation count', 'mean tag rank')
annoVsTagRanks.groupby('annotation count').mean().plot(ls='',marker='o',logx=True,alpha=0.5)
```

Activity Correlations - Tagging
---

```python
files = ('uniqueTagsPerUser','uniqueItemsPerUser','uniqueArtistsPerUser')
taggingData = annoPerUser
for fi in files:
    vars()['df_'+fi] = pd.read_table(d+fi,header=None,names=('uid','freq_'+fi),index_col='uid')
taggingData = taggingData.join([vars()['df_'+f] for f in files])
    
```

```python
fig,ax = subplots(nrows=1,ncols=1)

taggingData.columns = ['annotation count']+list(taggingData.columns[1:])
plotData = taggingData.groupby('annotation count').mean()
for fi in files:
    ax.plot(plotData.index,plotData['freq_'+fi],ls='',marker='o',label=fi)
ax.set_ylabel('n')
ax.set_xscale('log')
ax.set_yscale('log')
ax.legend(loc='upper left')
ax.set_xlabel('annotation count')
ax.set_title('mean activity for users with x annotations')

# Trying to figure out to use an exponentially weighted moving average, with little success
#pd.ewma(data.groupby('annotation count').mean(),com=100).plot(logx=True,ax=ax) 
```

Activity Correlations - Everything Else
---

```python
scrobblesPerItem = pd.read_table(d+'scrobblesPerItem',header=None,names=('iid','scrobbleCount'),index_col='iid')
annoPerItem = pd.read_table(d+'annotationsPerItem',header=None,names=('iid','annoCount'),index_col='iid')
data = scrobblesPerItem.join(annoPerItem)
```

```python
fig,ax = subplots(nrows=1,ncols=1)

plotData = data.groupby('annoCount').mean()
ax.plot(plotData.index,plotData['scrobbleCount'],ls='',marker='o')
ax.set_ylabel('number of scrobbles')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_xlabel('number of annotations')
ax.set_title('Mean number of scrobbles for items with x annotations')
```

```python
files = ('bannedPerUser','friendsPerUser','groupsPerUser','lovedPerUser')

data = annoPerUser
for fi in files:
    vars()['df_'+fi] = pd.read_table(d+fi,header=None,names=('uid','freq_'+fi),index_col='uid')
data = data.join([vars()['df_'+f] for f in files])
    
```

```python
fig,ax = subplots(nrows=1,ncols=1)

data.columns = ['annotation count']+list(data.columns[1:])
plotdata = data.groupby('annotation count').mean()
for fi in files:
    ax.plot(plotdata.index,plotdata['freq_'+fi].fillna(0),ls='',marker='o',label=fi)
ax.set_xlabel('annotation count')
ax.set_ylabel('n')
ax.set_xscale('log')
ax.set_yscale('log')
ax.set_title('mean activity for users with x annotations')
ax.legend()

```
