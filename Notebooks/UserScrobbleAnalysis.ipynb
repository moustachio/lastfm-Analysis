```python
### For starters, load in the data and prep as a Pandas data frame, formating all tracks to just indicate the corresponding artist

from urllib2 import unquote
import pandas as pd 
try: # just for copy/pasting without pylab
    pylab.rcParams['figure.figsize'] = (16.0, 12.0)
except:
    pass

# this function takes a track URL and returns a formatted artist name only
def artistFormat(st):
    if '+noredirect' in st:
        return unquote(st.strip().split('/')[1]).replace('+',' ')
    else:
        return unquote(st.strip().split('/')[0]).replace('+',' ')

# For now we'll just load a TSV of a particular user's data that we've dumped from the DB
#filename = 'hawkrobe.tsv'
filename = 'AxelStreichen.tsv'
#filename = 'jaimie.tsv'
#filename='data/11079838_scrobbles.txt'

data = pd.read_table(filename,header=None, names=('uid','trackURL','timestamp'), index_col='timestamp', parse_dates='timestamp') # Read in .tsv
data.pop('uid') # remove the UID column, since we only have one user
data.trackURL = data.trackURL.map(artistFormat)
#data = pd.get_dummies(data.trackURL.map(artistFormat))# this makes a unique column for each artist in the "artist" column
#data = data.sort() # ensure data is sorted by time

```

```python
# Calcualtes entropy of a frequency distribtion 
def ent(freqs):
	freqs = freqs[freqs>0]
	total = float(np.sum(freqs))
	freqs = freqs/total
	e = -np.sum(freqs*np.log2(freqs))
	n = len(freqs)
	p = 1.0/n
	eMax = - n*p*np.log2(p) # maximum entropy
	return e, e/eMax, n, total # returns entropy, relative entropy, unique items, total items

# Calcualtes gini coefficent of a frequency distribtion 
def gini(freqs):
    freqs = freqs[freqs>0]
    n = float(len(freqs))
    numSum = 0 # sum from numerator
    for i, freq in enumerate(sorted(freqs)):
        numSum += ((i+1)*float(freq)) 
    return (2*numSum)/(n*np.sum(freqs)) - ((n+1)/n)

# resample data monthly
monthly = data.resample('M', how='sum')

fig, axes = plt.subplots(nrows=2, ncols=2)

monthly.apply(lambda x: ent(x)[1],1).plot(ax=axes[0,0],label='rel. entropy') # plot entropy across artists each month
monthly.apply(gini,1).plot(ax=axes[0,0],label='gini coef.') # plot gini coef. across artists each month
axes[0,0].legend(loc='lower center')
axes[0,0].set_title('Gini coefficient and rel. entropy by month')

monthly.apply(lambda x: sum(x>0),1).plot(ax=axes[0,1])# total number of unique artists listened to each month
axes[0,1].set_title('Unique artists listened by month')

(monthly.cumsum()>0).apply(sum,1).plot(ax=axes[1,0]) # cumulative new artists over time
axes[1,0].set_title('Cumulative number of unique artists over time')

### I'll think of something else for the 4th axis
```
